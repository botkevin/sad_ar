```python
def render(img, obj, projection, position, scale=.01, color=False):
    vertices = obj.vertices
    scale_matrix = np.eye(3) * scale
    h, w, z = position

    for face in obj.faces:
        face_vertices = face[0]
        points = np.array([vertices[vertex - 1] for vertex in face_vertices])
        points = np.dot(points, scale_matrix)
        # render model in the middle of the reference surface. To do so,
        # model points must be displaced
        points = np.array([[p[0]+w, p[1]+h, p[2]+z] for p in points])
        dst = pc.warp_points(points, projection)
        imgpts = np.int32(dst)
        if color is False:
#             r = hash(np.asarray(face).tostring())
            r = int(np.sum(face[-1])**4)
            c = (r%243, r%179, r%52)
#             print (c)
            cv2.fillConvexPoly(img, imgpts, c)
        else:
            color = hex_to_rgb(face[-1])
            color = color[::-1] # reverse
            cv2.fillConvexPoly(img, imgpts, color)

    return img
```

```python
def render_mult (vid, obj, position, Hs):
    vid = vid[:-1]
    new_frames = []
    for fi in tqdm(range(len(vid))):
        frame = vid[fi].copy()
        H = Hs[fi]
        new_frame = render (frame, obj, H, position)
        new_frames.append (new_frame)
    return new_frames, Hs
```
