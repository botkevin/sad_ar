```python
from skvideo.io import vread, vwrite
import numpy as np
from matplotlib import pyplot as plt
import cv2
import csv
from tqdm import tqdm
```

```python
plt.rcParams['figure.figsize'] = 15, 15
```

```python
vid = vread ('data/box2.mp4')
```

```python
img0 = vid[0]
plt.imshow(img0)
```

```python
def add_points (points, img , n):
    plt.imshow(img)
    pts = plt.ginput(n,timeout=0)
    print (pts)
    plt.close()
    points += pts
```

```python
points = []
```

```python
%matplotlib qt
add_points (points, img0 , 20)
%matplotlib inline
plt.rcParams['figure.figsize'] = 15, 15
```

```python
with open('points2.txt', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile, delimiter=',', quotechar='|')
    for p in points:
        writer.writerow(p)

points = []

with open('points2.txt', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in spamreader:
        points.append(row)
points = np.asarray(points).astype(np.float64)
```

```python
len(points)
```

```python
def circle_frame (frame, points):
    frame = frame.copy()
    for p in points:
        frame = cv2.circle(frame, (int(p[0]),int(p[1])), 10, (0,255,0), 2)
    return frame
```

```python
img0p = circle_frame (img0, points)
plt.imshow(img0p)
plt.show()
```

```python
p_world = [(0,0,0), (4,0,0), (8,0,0), (12,0,0),
           (0,0,2), (4,0,2), (8,0,2), (12,0,2), (16,0,2),
           (0,0,4), (4,0,4), (8,0,4), (12,0,4), (16,0,4),
           (0,4,4), (4,4,4), (8,4,4), (12,4,4), (16,4,4),
           (0,8,4)]
assert len(points) == len(p_world)           
```

```python
# video = cv2.VideoCapture('data/box.avi')

# # Exit if video not opened.
# if not video.isOpened():
#     print ("Could not open video")
#     raise NotImplementedError

trackers = []
for i in range(len(points)):
    trackers.append(cv2.TrackerMedianFlow_create())

# Read first frame.
# ok, frame = video.read()
# if not ok:
#     print ('Cannot read video file')
#     raise NotImplementedError

frame = vid[0]

new_vid=[vid[0]]

box_len = 20
bboxes = []
for point in points:
    # Define an initial bounding box
    bbox = point - [box_len//2,box_len//2]
    bbox = np.concatenate((bbox, [box_len,box_len]))
    bbox = tuple(bbox.astype(int))
    bboxes.append(bbox)
# print (len(bboxes))

# Initialize tracker with first frame and bounding box
for i in range(len(points)):
    bbox = bboxes[i]
    tracker = trackers[i]
    ok = tracker.init(frame, bbox)
    if not ok:
        raise Exception

for frame in tqdm(vid[1:]):
    
    frame = frame.copy()

    # Start timer
    timer = cv2.getTickCount()

    # Update tracker
    oks = [True]*len(points)
    for i in range(len(points)):
        tracker = trackers[i]
        ok, bbox = tracker.update(frame)
        oks[i] = ok
        bboxes[i] = bbox

    # Calculate Frames per second (FPS)
    fps = cv2.getTickFrequency() / (cv2.getTickCount() - timer);

    # Draw bounding box
    for i in range(len(points)):
        ok = oks[i]
        if ok: 
            # Tracking success
            bbox = bboxes[i]
            p1 = (int(bbox[0]), int(bbox[1]))
            p2 = (int(bbox[0] + bbox[2]), int(bbox[1] + bbox[3]))
            cv2.rectangle(frame, p1, p2, (255,0,0), 2, 1)
        else :
            # Tracking failure
            cv2.putText(frame, "Tracking failure detected", (100,80), cv2.FONT_HERSHEY_SIMPLEX, 0.75,(0,0,255),2)

    # Display tracker type on frame
    cv2.putText(frame, " Tracker", (100,20), cv2.FONT_HERSHEY_SIMPLEX, 0.75, (50,170,50),2);

    # Display FPS on frame
    cv2.putText(frame, "FPS : " + str(int(fps)), (100,50), cv2.FONT_HERSHEY_SIMPLEX, 0.75, (50,170,50), 2);

    # Display result
    new_vid.append(frame)

    # Exit if ESC pressed
    k = cv2.waitKey(1) & 0xff
    if k == 27 : break

```

```python
vwrite("data/all_points2.mp4",np.asarray(new_vid))
```
